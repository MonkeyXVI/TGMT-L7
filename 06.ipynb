```python
# import lớp Image
from PIL import Image
from imgtool import *
from IPython.display import display # to display images
```

```python
# đường dẫn ảnh
my_path = 'G:\\thigiacmaytinh\\hinhanh\\04.jpg'
```

```python
# đọc ảnh 
img = load_image(my_path)

# Resize
# img = img.resize((250,250))
```

```python
display(img)
```

```python
# Chuyển từ ảnh RGB sang ảnh xám
gray_image = img.convert("L")
display(gray_image)
```

```python
# Phân RGB thành 3 tầng màu khác nhau
red_band, green_band, blue_band = img.split()
display(red_band)
display(green_band)
display(blue_band)
```

```python
# Trộn ảnh
merged_image =  Image.merge("RGB", (red_band, green_band, blue_band))
display(merged_image)
```
