```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

```

```python
# Đọc hình ảnh từ tệp
my_path = 'G:\\thigiacmaytinh\\hinhanh\\02.jpg'
image = Image.open(my_path)

# chuyển đổi ảnh thành mảng numpy
im_array = np.array(image.convert("L"))

# Biến đổi ảnh xám theo yêu cầu
inverted_image = 255 - im_array  # Invert ảnh bằng cách lấy 255 - giá trị pixel
squared_image = im_array ** 2  # Bình phương giá trị pixel
clamped_image = np.clip(im_array, 100, 200)  # Giới hạn giá trị pixel trong khoảng 100-200

# Hiển thị hình ảnh gốc và hình ảnh đã biến đổi với khoảng trắng
fig, axs = plt.subplots(2, 2, figsize=(10, 10))
fig.subplots_adjust(wspace=0.2, hspace=0.2)

axs[0, 0].imshow(im_array, cmap="gray")
axs[0, 0].set_title('Hình ảnh gốc')

axs[0, 1].imshow(inverted_image, cmap="gray")
axs[0, 1].set_title('Hình ảnh đã biến đổi (Inverted)')

axs[1, 0].imshow(squared_image, cmap="gray")
axs[1, 0].set_title('Hình ảnh đã biến đổi (Squared)')

axs[1, 1].imshow(clamped_image, cmap="gray")
axs[1, 1].set_title('Hình ảnh đã biến đổi (Clamped)')

plt.show()
```
