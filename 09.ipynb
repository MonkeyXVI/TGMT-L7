```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ng dn nh
my_path = 'G:\\thigiacmaytinh\\hinhanh\\03.jpg'
# c nh
im = Image.open(my_path)
plt.imshow(im)
```

```python
# chuyn sang nh xm
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# to mt hnh v mi
plt.figure()
plt.hist(im_array.flatten(), bins=6)
plt.xlabel('Gi tr ca pixel')
plt.ylabel('S lng')
plt.show()
```

```python
# lm ni bt ng bin
plt.figure()
plt.gray()
plt.contour(im_gray, orgin='image')
plt.show()
```
