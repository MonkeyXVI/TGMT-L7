```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Đường dẫn ảnh
my_path = 'G:\\thigiacmaytinh\\hinhanh\\03.jpg'
# Đọc ảnh
im = Image.open(my_path)
plt.imshow(im)
```

```python
# chuyển sang ảnh xám
im_gray = im.convert('L')
plt.imshow(im_gray, cmap='gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# tạo một hình vẽ mới
plt.figure()
plt.hist(im_array.flatten(), bins=6)
plt.xlabel('Giá trị của pixel')
plt.ylabel('Số lượng')
plt.show()
```

```python
# làm nổi bật đường biên
plt.figure()
plt.gray()
plt.contour(im_gray, orgin='image')
plt.show()
```

```python
# đường dân ảnh
my_path = 'G:\\thigiacmaytinh\\hinhanh\\03.jpg'
# đọc ảnh và hiển thị ảnh với trục tạo độ
im= Image.open(my_path).convert('L')

# hiển thị ảnh trên trục tạo độ
plt.imshow(im)

# tạo một ình vẽ mới
plt.figure()

# đặt màu sắc là xám
plt.gray()

# vẽ biểu đồ đường biên (contour) của ảnh
plt.contour(im, origin="image")

# đảm bảo tỷ lệ trục x và trục y trên biểu đồ là bằng nhau
plt.axis('equal')

# hiển thị ảnh
plt.show()
```
