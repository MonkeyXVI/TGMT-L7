```python
from PIL import Image
from imgtool import *
```

```python
#Đường dẫn ảnh
my_path = 'G:\\thigiacmaytinh\\hinhanh\\04.jpg'
```

```python
# Doc anh
img = load_image(my_path)
```

```python
display(img)
```

```python
# khu vự cắt
region = (100 , 75, 175, 150)
# Cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
# Dán 
# Xác định vị cần dán
past_position = (100,500)
img.paste(cropped_image, past_position)
display(img)
```

```python
#thay đổi kích thước
thumbnail_size= (50,50)
img.thumbnail(thumbnail_size)
display (img)
```
