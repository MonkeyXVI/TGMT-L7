```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
from PIL import Image

# c nh
image = Image.open('G:\\thigiacmaytinh\\hinhanh\\10.jpg')

# Chuyn sang nh xm
gray_image = image.convert('L')

image_array = np.array(gray_image)

# Chuyn sang nh nh phn
binary_image = 1*(image_array<200)

# Thc hin opening  loi b nhiu v ni cc i tng
opened_image = ndimage.binary_closing(binary_image, structure=np.ones((5,5)))

# nh nhn
labeled_image, num_features = ndimage.label(opened_image)

# To li subplot vi 3 dng v 1 ct
fig, axs = plt.subplots(4, 1, figsize=(50, 50))

plt.subplot(4,1,1)
plt.imshow(gray_image, cmap='gray')
plt.title("nh xm")

plt.subplot(4,1,2)
plt.imshow(binary_image, cmap='gray')
plt.title("nh nh phn")
                                      
plt.subplot(4,1,3)
plt.imshow(opened_image, cmap='gray')
plt.title("nh sau opening")


plt.subplot(4,1,4)
plt.imshow(labeled_image, cmap='gray')
plt.title(f" nh nhn ({num_features} i tng)")

plt.show();
```
